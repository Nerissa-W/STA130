The key factor that distinguishes ideas that can be examined and tested statistically from those that cannot is measurability. For an idea to be testable, it must be possible to collect data that can be measured and analyzed. This means the idea should be specific enough to allow for the collection of quantitative data. A good null hypothesis is one that is clear, precise, and testable. It should state that there is no effect or no difference, providing a baseline that can be challenged by the data. For example, The average height of adult men in a city is 170 cm is a good null hypothesis because it is specific and measurable. In the context of hypothesis testing, the null hypothesis (( H_0 )) represents the default position that there is no effect or no difference, while the alternative hypothesis (( H_A )) represents the position that there is an effect or a difference. The purpose of hypothesis testing is to determine whether there is enough evidence in the sample data to reject the null hypothesis in favor of the alternative hypothesis. This process helps researchers make informed conclusions about the population based on sample data.

**Some terminology**
- ( x_i ): These are the individual data points in your sample. For example, if you surveyed 10 people about their heights, each persons height is an ( x_i ).
- ( \bar{x} ): This is the average (mean) of all the ( x_i ) values in your sample. So, if you add up all the heights and divide by 10, you get ( \bar{x} ), the sample average height.
- ( \mu ): This is the true average (mean) of the entire population youre interested in. If you could measure the height of every single person in the population, ( \mu ) would be the average height.
- ( \mu_0 ): This is the hypothesized average value of the population under the null hypothesis. For example, ( H_0: \mu = \mu_0 ) might state that the average height of the population is 170 cm.

**Hypothesis Testing**
- Null Hypothesis (( H_0 )): This is a statement that there is no effect or no difference. For example, ( H_0: \mu = \mu_0 ) means we assume the population average (( \mu )) is equal to a specific value (( \mu_0 )).
- Alternative Hypothesis (( H_A )): This is what you want to prove. It states that the population average (( \mu )) is not equal to the hypothesized value (( \mu_0 )).

In hypothesis testing, we use the sample data (( x_i ) and ( \bar{x} )) to make inferences about the population parameter (( \mu )). We compare the sample average (( \bar{x} )) to the hypothesized population average (( \mu_0 )) to see if there is enough evidence to reject the null hypothesis (( H_0 )). If we reject ( H_0 ), we conclude that the true population mean (( \mu )) is different from the hypothesized mean (( \mu_0 )).

**Simplified Explanation**

Imagine you taste a spoonful of soup (your sample) to decide if the whole pot (the population) needs more salt. The taste of the spoonful ( ( \bar{x} ) ) helps you decide about the whole pot ( ( \mu ) ). The test result is about the whole pot, not just the spoonful.

When we calculate a p-value, we start by assuming that the null hypothesis is true. This means we imagine a world where there is no effect or no difference, and any observed data is due to random chance. By doing this, we can determine how likely it is to observe our sample data, or something more extreme, under this assumption. If the p-value is very small, it suggests that our observed data is highly unlikely in this imagined world where the null hypothesis is true. This gives us strong evidence to reject the null hypothesis in favor of the alternative hypothesis, which proposes that there is an effect or a difference. Essentially, imagining the null hypothesis as true provides a baseline to compare our actual data against, helping us understand whether our results are due to random chance or if they indicate a real effect.

When we talk about p-values in hypothesis testing, a smaller p-value indicates that the observed data is very unlikely under the null hypothesis. The p-value represents the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed, assuming the null hypothesis is true. If this probability is very low, it suggests that the observed data is not consistent with the null hypothesis. For example, if we hypothesize that a coin is fair (null hypothesis) and then flip it 100 times, getting 90 heads would result in a very small p-value. This small p-value implies that such an extreme result is highly improbable if the coin were truly fair, making the null hypothesis (that the coin is fair) seem implausible or ridiculous. Essentially, the smaller the p-value, the stronger the evidence against the null hypothesis, indicating that the observed data is unlikely to occur by random chance alone if the null hypothesis were true. This helps us decide whether to reject the null hypothesis in favor of the alternative hypothesis.

In the second half of the first pre-lecture video, the concept of a one-sided (or one-tailed) test is introduced as a way to test a hypothesis in a specific direction, as opposed to a two-sided (or two-tailed) test, which tests for differences in both directions. To adjust the code from Demo II of the Week 5 TUT for a one-sided hypothesis test, we need to modify the calculation of the p-value. In a two-sided test, the p-value is calculated by considering the extreme values in both tails of the distribution. However, in a one-sided test, we only consider the extreme values in one tail, depending on the direction of the hypothesis.

For example, if the original code for a two-sided test calculates the p-value as 2 * min(p-value from one tail, p-value from the other tail), we would change it to simply use the p-value from the relevant tail for a one-sided test. This change means that the p-value for a one-sided test will generally be smaller than for a two-sided test, as we are only considering one direction of extreme values.

This change in the code affects the interpretation of the hypothesis test by focusing on whether the observed data provides strong evidence against the null hypothesis in a specific direction. For instance, if we are testing whether a vaccine is more effective than a placebo, a one-sided test would only consider the possibility that the vaccine is more effective, not less effective. Consequently, we should expect the p-value to be smaller in a one-sided test compared to a two-sided test, as it is easier to find significant results when only one direction is considered. This makes one-sided tests more powerful in detecting an effect in a specified direction, but they should be used with caution to avoid bias.

# Problem Introduction

This project analyzes an experiment with 80 STA130 students, where 49 students correctly identified whether milk or tea was poured first.
The goal is to determine if the proportion of students who can correctly identify which was poured first is greater than what would be expected by random guessing (0.5).

### Relationship with Original Experiment

The original experiment involved Fisher and Dr. Muriel Bristol, where Bristol claimed she could taste the difference between tea poured first and milk poured first. Fisher designed a hypothesis test to examine this claim. In this project, we conduct a similar experiment with STA130 students to test their ability to identify the order of pouring.

### Statements of the Null Hypothesis and Alternative Hypothesis

- Null Hypothesis (H): The proportion of STA130 students who can correctly identify which was poured first is equal to 0.5 (random guessing).
- Alternative Hypothesis (H): The proportion of STA130 students who can correctly identify which was poured first is greater than 0.5.

### Quantitative Analysis

```python
# Quantitative Analysis
def quantitative_analysis():
    # Sample size
    n = 80

    # Number of successes (students who correctly identified which was poured first)
    x = 49

    # Sample proportion
    p_hat = x / n

    # Null hypothesis proportion
    p_0 = 0.5

    # Standard error
    se = (p_0 * (1 - p_0) / n) ** 0.5

    # Z-score
    z = (p_hat - p_0) / se

    # P-value (one-tailed test)
    p_value = 1 - stats.norm.cdf(z)

    print(f"Sample size (n): {n}")
    print(f"Number of successes (x): {x}")
    print(f"Sample proportion (p): {p_hat:.4f}")
    print(f"Null hypothesis proportion (p): {p_0}")
    print(f"Standard error (SE): {se:.4f}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")

    return p_value, z

p_value, z = quantitative_analysis()
```

### Methodology Code and Explanations

We use a one-sample z-test for proportions to determine if the observed proportion of students who correctly identified which was poured first is significantly greater than the proportion expected by random guessing (0.5).
The z-score measures how many standard errors the observed proportion is away from the null hypothesis proportion.
The p-value indicates the probability of obtaining a result at least as extreme as the observed result, assuming the null hypothesis is true.

### Supporting Visualization

```python
import scipy.stats as stats
import matplotlib.pyplot as plt


# Supporting Visualizations (if needed/optional)
def supporting_visualizations(z):
    plt.figure(figsize=(10, 6))
    
    # Plotting the normal distribution curve
    x_values = [i/100 for i in range(-400, 400)]
    y_values = [stats.norm.pdf(x) for x in x_values]
    
    plt.plot(x_values, y_values, label='Normal Distribution')
    
    # Marking the z-score on the plot
    plt.axvline(z, color='r', linestyle='--', label=f'Z-score: {z:.4f}')
    
    plt.title('Normal Distribution Curve with Z-score')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    
    plt.legend()
    
    plt.show()


supporting_visualizations(z)
```

### Findings and Discussion

```python
# Findings and Discussion
def findings_and_discussion(p_value):
    print("\nFindings and Discussion")
    
    # Sample size
    n = 80

    # Number of successes (students who correctly identified which was poured first)
    x = 49

    # Sample proportion
    p_hat = x / n

    # Null hypothesis proportion
    p_0 = 0.5

    # Standard error
    se = (p_0 * (1 - p_0) / n) ** 0.5

    # Z-score
    z = (p_hat - p_0) / se

    print(f"The sample size is {n}, and the number of successes is {x}. The sample proportion is {p_hat:.4f}.")
    print(f"The null hypothesis proportion is {p_0}, and the standard error is {se:.4f}. The calculated z-score is {z:.4f}, and the p-value is {p_value:.4f}.")
    if p_value < 0.05:
        print("Since the p-value is less than 0.05, we reject the null hypothesis (H). There is evidence that the proportion of STA130 students who can correctly identify which was poured first is greater than 0.5.")
    else:
        print("Since the p-value is not less than 0.05, we fail to reject the null hypothesis (H). There is not enough evidence to conclude that the proportion of STA130 students who can correctly identify which was poured first is greater than 0.5.")

findings_and_discussion(p_value)
```

### Ethical Professionalism Considerations
    
Using p-values and hypothesis testing appropriately is an important ethical and professional responsibility of anyone doing data analysis. Actually, there is quite the quiet Contra-Versy (or is it Con-TROV-ersy?) around p-values. First, on a general level, it seems quite clear that p-values and hypothesis testing methodologies MUST play some ongoing contributing role in the so-called "replication crisis" rampantly afflicting mordern science; namely, "significant findings" made in scientific studies are not able to be reproduced by future studies at an alarming rate; and, this whole paradigm of "significant findings" is based on p-values and hypothesis testing... so, something's going on with this methodology in some way...
    
More specifically however, p-values are themselves quite problematic. To see this, just briefly consider the following article titles...

- [Why are p-values controversial?](https://www.tandfonline.com/doi/full/10.1080/00031305.2016.1277161) 
- [What a nerdy debate about p-values shows about science and how to fix it](https://www.vox.com/science-and-health/2017/7/31/16021654/p-values-statistical-significance-redefine-0005)
- [The reign of the p-value is over: what alternative analyses could we employ to fill the power vacuum?](https://royalsocietypublishing.org/doi/10.1098/rsbl.2019.0174)
- [Scientists rise up against statistical significance](https://www.nature.com/articles/d41586-019-00857-9)
- [Statistics experts urge scientists to rethink the p-value](https://www.spectrumnews.org/news/statistics-experts-urge-scientists-rethink-p-value)

While the issues here are relatively advanced and subtle (as introduced [here](https://www2.stat.duke.edu/~berger/p-values.html), presented [here](https://www.jarad.me/courses/stat587Eng/slides/Inference/I06-Pvalues/why_pvalues_dont_mean_what_you_think_they_mean.pdf), and demonstrated using simulation [here](https://jaradniemi.shinyapps.io/pvalue/)), the problem essentially comes down to the fact that most scientists (or just people) don't know how to really interpret the numeric value of a p-value. There are therefore two current proposed solutions to address this challenge.
    
1. Just interpreting p-values using the follwing table (which really isn't that hard, so it's surprising that this solution isn't more broadly adopted...)
    
|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    

2. Only do **hypothesis testing** on the basis of confidence intervals, not **p-values** (which might be the best solution wherever doing so is a realistic, convenient  possibility...)

With this quite broad introductory context in mind, what does your favorite ChatBot thinks about the following statements? 
    
1. Hypothesis testing is not a "mathematical proof"<br><br>

    1. We do not prove $H_0$ false, we instead give evidence against the $H_0$: "We reject the null hypothesis with a p-value of XYZ, meaning we have ABC evidence against the null hypothesis"
    2. We do not prove $H_0$ is true, we instead do not have evidence to reject $H_0$: "We fail to reject the null hypothesis with a p-value of XYZ"<br><br>

2. Implying that a "non-significant result" means there is "no effect" misleads an audience because this may in actual fact simply indicate that there was insufficient evidence to reject the null hypothesis. So this therefore overlooks the possibility of sample size limitations, or Type II errors (which means a test incorrectly concludes that there is no effect or difference when, in fact, there is one). 
    
> Similarly, analagously, a "significant result" used to reject the null hypothsis could alternatively be a Type I error (which means a test actually incorrectly rejected a null hypothesis when it was actually true)... we're only providing a measure of evidence against the null hypothesis... but the evidence could still incorrectly suggest the wrong conclusion... it really depends on how strong the evidence is...
>
> - all of which is why just interpreting p-values using the table above is a good idea...

3. The p-values used for hypothesis testing are contructed upone the assumptions of the null hypotheses they correspond to; but, null hypotheses are actually often presented in simple forms that routinely hide a lot of information that is implicitly used to construct the p-values. For example, distributional assumptions about the population, estimated "plug-in" values that can used to simplify the problem calculations, and the reliance upon "random sampling", etc...<br><br>
           
4. Drawing overly broad conclusions, or making recommendations based on findings that reject the null hypothesis in a specific context is fraught with the problematic risks of overgeneralization errors. Further exacerbating this issue, null hypotheses are typically so called "point null hypotheses" which is meant to emphasize that they are mathematically increadibly sharply specific; whereas, alternative hypotheses are usually very unspecific. An alternative hypothesis that "the null hypothesis is false" doesn't say much... we should wonder, "how, specfically, is the null false?"
    
As an example really giving a demonstrating this, consider rejecting a null hypothesis that there is no correlation between rain and pizza's delivered. Such a decision doesn't specify what the actual hypothetical correlation might be. In fact, it doesn't even indicate if there are more or less pizzas delivered when it rains... 

> which, actually, shows very clearly why statistical inference using hypothesis testing is inferior to statistical inference based on confidence intervals...
> 
> - a confidence interval provides a range of plausible values of what the parameter in question might be; whereas, ...
> - trying to more clearly address what the plausible values of the parameter in question might be on the basis of hypothesis testing would require conducting further experiements to continously reject increasingly detailed hypothesies to narrow down what the alternative hypothesis might actually include... which would indeed be an utterly vapid misuse of the intended purpose of hypothesis testing entrprise...
